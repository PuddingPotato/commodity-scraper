```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

```python
url = 'https://www.dailymetalprice.com/metalpricecharts.php?c=st&u=mt&d=240'
response1 = requests.get(url)
```

```python
if response1.status_code == 200:
    print("success")
    page_html = BeautifulSoup(response1.text, 'html.parser')
else:
    print('not success')
```

```python
table = page_html.find('table', class_ = "table table-striped success table-hover table-condensed")
tbody = table.find('tbody')
row = tbody.find_all('tr')
```

## Steel Rebar

```python
a_tag = row[18].find('a', string = 'Steel Rebar')

#if a_tag:
price_td = a_tag.find_next('td')
price = price_td.text.strip().split()[0]
price = [price[1:]]

date_td = price_td.find_next('td')
date_span = date_td.find('span', class_ = 'hidden-md')
date = [date_span.text.strip()]
```

```python
rebar_df = pd.DataFrame({"วันที่" : date,
                         "ราคา (mt)" : price})
```

```python
rebar_df
```
